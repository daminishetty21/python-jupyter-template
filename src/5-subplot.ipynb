### Exercise 5: Subplot Layout

```python
#
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# First subplot: Line chart
axs[0, 0].plot(x, y, color='blue')
axs[0, 0].set_title('Line Chart')

# Second subplot: Bar chart
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axs[0, 1].set_title('Bar Chart')

# Third subplot: Scatter plot
axs[1, 0].scatter(x, y, color='green')
for i, (x_val, y_val) in enumerate(zip(x, y)):
    axs[1, 0].annotate(f"({x_val}, {y_val})", (x_val, y_val), textcoords="offset points", xytext=(0, 10), ha='center')
axs[1, 0].set_title('Scatter Plot')

# Fourth subplot: Pie chart
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
axs[1, 1].set_title('Pie Chart')

# Adjust layout to avoid overlap
plt.tight_layout()

# Show the plot
plt.show()
```
